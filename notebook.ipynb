```python
import os

file_weather = "weather.clean"
file_football = "football.clean"

if not os.path.isfile(file_weather):
    print(f"Weather file missing! Expected at: {file_weather}")
    
if not os.path.isfile(file_football):
    print(f"Weather file missing! Expected at: {file_football}")
```

```python
import pandas as pd
```

```python
df_football = pd.read_csv(file_football, sep=' ')
print(f"Number of rows: {len(df_football.index)}")
df_football.head()
```

```python
df_weather = pd.read_csv(file_weather, sep=' ')
print(f"Number of rows: {len(df_weather.index)}")
df_weather.head()
```
